```python
df_close_loan = pd.read_csv('datasets/D_close_loan.csv')
df_close_loan
```

```python
df_loan = pd.read_csv('datasets/D_loan.csv')
df_loan
```

```python
merged_loans = pd.merge(df_loan, df_close_loan, how='inner', on = 'ID_LOAN')
```

```python
merged_loans
```

```python
total_loans = merged_loans.groupby('ID_CLIENT').size().reset_index(name='LOAN_NUM_TOTAL')
total_loans
```

```python
closed_loans = merged_loans[merged_loans['CLOSED_FL'] == 0].groupby('ID_CLIENT').size().reset_index(name='LOAN_NUM_CLOSED')
closed_loans
```

```python
merged_data = merged_data.merge(total_loans, how='left', left_on='ID', right_on='ID_CLIENT')
```

```python
merged_data = merged_data.merge(closed_loans, how='left', left_on='ID', right_on='ID_CLIENT')
merged_data
```

```python
merged_data = merged_data.drop(columns =['ID_CLIENT_x', 'ID_CLIENT_y']).fillna(0)
merged_data
```

Проверим, что нет ошибки в разнице между всеми и закрытыми кредитами

```python
assert (merged_data[merged_data['LOAN_NUM_CLOSED'] > merged_data['LOAN_NUM_TOTAL']].empty)
```

```python
df_target = pd.read_csv('datasets/D_target.csv')
df_target
```

```python
df_target = df_target.merge(merged_data, how='left', left_on='ID_CLIENT', right_on='ID')
df_target
```

```python
df_target.drop(columns = ['ID_CLIENT', 'ID'], inplace=True)
```
